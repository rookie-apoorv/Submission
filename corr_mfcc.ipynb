```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
def corr_matrix_mfcc(df) :
    df = df.T
    corre_matrixx = df.corr()
    sns.heatmap(corre_matrixx, cmap = 'coolwarm')
    return corre_matrixx
```

```python
file1 =  pd.read_excel('Try_1_heatmap.xlsx', header = None)
matrix_ineed = corr_matrix_mfcc(file1)
```

```python
file35 = pd.read_csv('master_file_umap_1_(18).csv', header = None)
file35 = file35.drop(index = 0).drop(columns = file35.columns[0])
corr_matrix_mfcc(file35)
```
