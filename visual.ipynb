```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the UMAP features
file_path = 'master_cluster_umap_1_(2).csv'  # Replace with your UMAP features CSV file
data = pd.read_csv(file_path)

# Extract UMAP components
umap_1 = data['UMAP_1']  # Column for the first UMAP component
umap_2 = data['UMAP_2']  # Column for the second UMAP component
print(data.columns)
# Optional: Extract cluster labels if available
# Uncomment the line below if you have a 'cluster' column in your data
labels = data['cluster']

# Plot the UMAP features
plt.figure(figsize=(10, 8))
# plt.scatter(umap_1, umap_2, s=30, alpha=0.7)  # Change 's' for point size, 'alpha' for transparency

# Optional: If you have labels, color by label
# Replace 'c=labels' if you want to use labels for color-coding
plt.scatter(umap_1, umap_2, c=labels, cmap='Set1', s=30, alpha=0.7)

plt.xlabel('UMAP Dimension 1')
plt.ylabel('UMAP Dimension 2')
plt.title('2D UMAP Projection')
plt.colorbar(label='Cluster Label')  # Use if color-coding with labels

plt.show()
```
