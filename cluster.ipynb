```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
file_path = 'master_nat_1_umap_(18).csv'
data = pd.read_csv(file_path)
X = data.iloc[:,1:19].values
file_names = data.iloc[:,0]
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)
n_clusters = 6
kmeans  = KMeans(n_clusters = n_clusters, random_state= 42)
label_predict = kmeans.fit_predict(X_normalized)
predict = pd.DataFrame()
predict = data
predict['cluster'] = label_predict
predict.to_csv('master_cluster_nat_umap_1_(18).csv')


```

```python
import matplotlib.pyplot as plt
inertias = []
for i in range(1,10) :
    n_clusters = i
    kmeans  = KMeans(n_clusters = n_clusters, random_state= 42)
    kmeans.fit(X_normalized)
    inertias.append(kmeans.inertia_)
plt.figure(figsize = (12,8))
plt.plot(range(1,10), inertias ,marker = 'o')
plt.show()
```
