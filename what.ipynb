```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
num_files = 50 
cols = 5
rows = (num_files // cols) + (num_files % cols > 0)
plt.figure(figsize = (20, 4*rows))
for i in range (1, num_files + 1) :
    csv_file = f"{i:02d}-MFCC.csv"

    data = pd.read_csv(csv_file, header = None)

    data = data.drop(index = 0)
    plt.subplot(rows, cols, i)
    sns.heatmap(data)
    plt.title(csv_file)
plt.tight_layout()
```

```python
def num_seconds(file_name) :
    with open(f'{file_name}','r') as file:
        first_line = file.readline()
        num_columns =  len(first_line.split(','))
        return num_columns/86
```

```python
size = pd.DataFrame(columns=['File', 'Time'])
for i in range(1,116) :
    file_name = f'{i:02d}-MFCC.csv'
    time_value = num_seconds(file_name)
    size.loc[i] = [file_name, time_value]
```
