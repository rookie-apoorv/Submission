```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

file_path = 'train_master_file_pca.csv'
data = pd.read_csv(file_path)
X = data.iloc[:,1:19].values
file_names = data.iloc[:,0]
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)
n_clusters = 6
kmeans  = KMeans(n_clusters = n_clusters, random_state= 42)
label_predict = kmeans.fit_predict(X_normalized)
predict = pd.DataFrame()
predict = data
predict['cluster'] = label_predict
predict.to_csv('train_master_file_pca_cluster.csv')
```

```python

from sklearn.metrics import silhouette_score,calinski_harabasz_score, davies_bouldin_score

# Load the CSV file
data = pd.read_csv("train_master_cluster_umap.csv")

# Extract features and labels
features = data.iloc[:, 2:-1]  # All columns except the first two and last
labels = data.iloc[:, -1]  # Last column as labels

# Calculate the silhouette score
score = silhouette_score(features, labels)
calinski_harabasz = calinski_harabasz_score(features, labels)
davies_bouldin = davies_bouldin_score(features, labels)
print(f'Silhouette Score: {score}')
print(f'Calinski-Harabasz Index: {calinski_harabasz}')
print(f'Davies-Bouldin Index: {davies_bouldin}')
```
