```python
import pandas as pd
import scipy.cluster.hierarchy as sch
import seaborn as sns
import matplotlib.pyplot as plt

# Load the correlation matrix
file_path = 'corr_matrix_to_sort.csv'
corr_matrix = pd.read_csv(file_path, index_col=0)

# Perform hierarchical clustering
linkage = sch.linkage(corr_matrix, method='ward', metric = 'euclidean')

# Get the order of the clustered columns
dendro = sch.dendrogram(linkage, no_plot=True)
clustered_order = dendro['leaves']

# Reorder the correlation matrix based on the clustering
sorted_corr_matrix = corr_matrix.iloc[clustered_order, clustered_order]

# Plot the reordered correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(sorted_corr_matrix, cmap='plasma', vmin=0.75, vmax=1)
plt.title('Clustered Correlation Matrix')
plt.show()
```

```python
sorted_corr_matrix.to_csv('sorted.csv')
```
